```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
def building_oscillation(_t, y, c):
    x, v = y
    dxdt = v
    dvdt = -(c / 10 ** 5) * v - (10 ** 6 / 10 ** 5) * x
    return [dxdt, dvdt]
```

```python
c_values = [0, 1e5]
x0 = [1, 0]

t = np.linspace(0, 10, 1000)
```

```python
plt.figure(figsize=(10, 6))
for c in c_values:
    solution = solve_ivp(
        building_oscillation,
        [t[0], t[-1]],
        x0,
        t_eval=t,
        args=(c,)
    )
    plt.plot(solution.t, solution.y[0], label=f"c = {c:.0e}")
    plt.title("Oppgave 2a")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
```
