```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f1(t, _):
    # c = 10^5
    c1 = 1
    c2 = 0
    i = 1j
    constant = (np.sqrt(39) * i) / 2
    return np.exp(-0.5 * t) * (c1 * np.cos(constant * t) + c2 * np.sin(constant * t))

def f2(t, _):
    # c = 0
    c1 = 1
    c2 = 0
    constant = np.sqrt(10)
    return c1 * np.cos(constant * t) + c2 * np.sin(constant * t)
```

```python
t0 = 0
t_end = 10
x0 = 1
dxdt0 = 0
t = np.linspace(t0, 3, 100) 
sol1 = solve_ivp(f1, [t0, t_end], [x0], t_eval=t)
sol2 = solve_ivp(f2, [t0, t_end], [x0], t_eval=t)
```

```python
fig, ax1 = plt.subplots()

ax1.set_title("Oppgave 2a")
ax1.plot(sol1.t, sol1.y[0], label="$c=10^5$", color="blue")
ax1.set_ylabel("y-akse for $c=10^5$")
ax1.tick_params(axis="y")
ax1.legend(loc="upper left")

ax2 = ax1.twinx()
ax2.plot(sol2.t, sol2.y[0], label="$c=0$", color="red")
ax2.set_ylabel("y-akse for $c=0$")
ax2.tick_params(axis="y")
ax2.legend(loc="upper right")
```
